```python
x1 = random.expovariate(1) # Generate a random number using exponential distribution
values1 = [ random.expovariate(1) for i in range(1000) ]
plt.plot(values1)
```

```python
# random choice from list
greetings = ['hello', 'hi', 'hey', 'hola', 'bonjour']
random.choice(greetings) # random choice from list
random.choices(greetings, k=7) # random choice from list with k=3
```

```python
# weighted random choice from list
colors = ["red", "green", "blue", "yellow", "orange", "purple"]
results = random.choices(colors, weights=[5.2, 0.1, 0.3, 0.2, 0.2, 0.2], k=10)
results
```

```python
# Shuffle a list
deck = list(range(1,53))
random.shuffle(deck)
print(deck)
```

```python
# randome sample from list
sample = random.sample(deck, k=5)
print(sample)
```

```python
# Generating fake data 
first_names = ["Alice", "Bob", "Cathy", "Dan", "Ed", "Frank", "George"]
last_names = ["Smith", "Jones", "Brown", "Davis", "Garcia", "Miller"]

cities = ["New York City", "Los Angeles", "Chicago", "Houston", "Phoenix", "Philadelphia", "San Antonio", "San Diego", "Dallas", "San Jose"]

states = ["New York", "California", "Illinois", "Texas", "Arizona", "Florida", "Pennsylvania", "California", "Illinois", "Texas"]

data = []
for i in range(10):
    first = random.choice(first_names)
    last = random.choice(last_names)
    
    phone = f"{random.randint(100,999)}-{random.randint(100,999)}-{random.randint(1000,9999)}"
    
    street = random.randint(100,999)
    city = random.choice(cities)
    state = random.choice(states)
    zipcode = random.randint(10000, 99999)
    
    address = f"{street} {city} {state} {zipcode}"
    
    email = first.lower() + "." + last.lower() + "@gmail.com"
    print(f"{first} {last} {phone} {address} {email}\n")
    data.append([first, last, phone, address, email])
```
